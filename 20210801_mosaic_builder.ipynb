```python
from mosaic import MosaicGrid
import numpy as np
%load_ext autoreload 
%autoreload 2
%config Completer.use_jedi = False
```

```python
# main object
mg = MosaicGrid(
    image_dir='selected_images',
    n_rows=11,
    n_cols=7,
    pixels_per_block_x=128,
    pixels_per_block_y=128,
    glob_str='*.png'
)
```

```python
# see files in selected directory
mg.files[:5]
```

## add crowd images

take up 3 blocks each

```python
mg.add_image('r003.png', start_row=0, start_col=0, size=3, invert=False)
mg.add_image('r002.png', start_row=8, start_col=4, size=3, invert=False)
```

```python
mg.show_mosaic()
```

## add big smileys

takes up 2 blocks

```python
mg.add_image('s105.png', start_row=0, start_col=5, size=2, invert=True)

mg.add_image('s103.png', start_row=4, start_col=0, size=2, invert=True)

mg.add_image('s107.png', start_row=8, start_col=0, size=2, invert=True)

mg.add_image('s101.png', start_row=5, start_col=3, size=2, invert=True)

mg.add_image('s106.png', start_row=3, start_col=5, size=2, invert=True)
```

```python
mg.show_mosaic()
```

## add remaining smileys

```python
remaining_smileys = [f.name for f in mg.files if f.name[0] not in ['s', 'r']]
```

```python
for ii, (irow, icol) in enumerate(mg.unfilled_block_coords):
    mg.add_image(remaining_smileys[ii], start_row=irow, start_col=icol, invert=True)
```

```python
mg.show_mosaic()
```
